```python
import pandas as pd
import numpy as np
import math

# remove corrupted data

df = pd.read_csv('train.csv')
# del df['MOFname']
df.replace(0, np.nan, inplace=True)
df.replace(-1, np.nan, inplace=True)
df.replace(math.inf, np.nan, inplace=True)
df.dropna(axis=0, how='any', inplace=True)
df.to_csv('train_rm.csv')
df
```

```python
mean = df.mean(axis=0)
median = df.median(axis=0)
```

```python
# mean (exclude nan at functional_groups and topology)
# TODO word embedding at functional_groups and topology to improve this method

df = pd.read_csv('train.csv')
df.dropna(subset=['functional_groups', 'topology'], inplace=True)
df.replace(0, np.nan, inplace=True)
df.replace(-1, np.nan, inplace=True)
df.replace(math.inf, np.nan, inplace=True)
df.fillna(mean, inplace=True)
df.to_csv('train_mean.csv')
df
```

```python
# median (exclude nan at functional_groups and topology)
# TODO word embedding at functional_groups and topology to improve this method

df = pd.read_csv('train.csv')
df.dropna(subset=['functional_groups', 'topology'], inplace=True)
df.replace(0, np.nan, inplace=True)
df.replace(-1, np.nan, inplace=True)
df.replace(math.inf, np.nan, inplace=True)
df.fillna(median, inplace=True)
df.to_csv('train_median.csv')
df
```
